   This dataset was generously provided by Olist, the marketplaces which performed since 2015 as SaaS technology company in Brazilian. Olist connects small businesses from all over Brazil to channels without hassle and with a single contract. Those merchants are able to sell their products through the Olist Store and ship them directly to the customers using Olist logistics partners.
   
**The dataset information of 100k orders from 2016 to 2018:**

>- Customers Dataset = This dataset has information about the customer and its location.
>- Geolocation Dataset = This dataset has information Brazilian zip codes and its lat/lng coordinates
>- Order Items Dataset = This dataset includes data about the items purchased within each order.
>- Payments Dataset = This dataset includes data about the orders payment options.
>- Order Reviews Dataset = This dataset includes data about the reviews made by the customers.
>- Order Dataset = This is the core dataset. From each order you might find all other information.
>- Products Dataset = This dataset includes data about the products sold by Olist.
>- Sellers Dataset = This dataset includes data about the sellers that fulfilled orders made at Olist.
>- Category Name Translation = Translates the product_category_name to english.

### Read

```python
df= pd.read_csv('data_kotor.csv')
df.head()
```

### Drop

```python
df['product_photos_qty'].fillna(value= 0, inplace= True)
df['product_description_lenght'].fillna(df['product_description_lenght'].median(), inplace= True)
df['product_name_lenght'].fillna(df['product_name_lenght'].mode()[0], inplace= True)
df['product_category_name'].fillna(value= 'Uncategorized', inplace= True)
df['product_weight_g'].fillna(df['product_weight_g'].mode()[0], inplace= True)
df['product_length_cm'].fillna(df['product_weight_g'].mode()[0], inplace= True)
df['product_height_cm'].fillna(df['product_height_cm'].mode()[0], inplace= True)
df['product_width_cm'].fillna(df['product_width_cm'].mode()[0], inplace= True)
df['payment_value'].fillna(df['payment_value'].mean(), inplace = True)
df['payment_installments'].fillna(df['payment_installments'].mode()[0], inplace= True)
df['payment_type'].fillna(df['payment_type'].mode()[0], inplace= True)
df['payment_sequential'].fillna(df['payment_sequential'].mode()[0], inplace= True)

df= df.drop(['review_comment_title', 'review_comment_message', 'order_id','order_item_id','customer_id',
             'seller_id','product_id','review_id','customer_unique_id','product_name_lenght',
             'product_description_lenght','product_photos_qty','product_length_cm',
             'product_height_cm','product_width_cm','seller_zip_code_prefix', 
             'customer_zip_code_prefix', 'order_estimated_delivery_date',  
              'payment_sequential'], axis= 1)
```

```python
df.head()
```

### Translate

```python
cat_trans = pd.read_csv("product_category_name_translation.csv")
cat_trans
```

### Joining Data

```python
df= pd.merge(df, cat_trans, on='product_category_name', how='left')
df.head()
```

```python
df['product_category_name_english'].fillna(value= 'Uncategorized', inplace= True)
df['product_category_name_english'].isnull().sum()
```

```python
df.dropna(subset= ['order_approved_at', 'order_delivered_customer_date', 'order_delivered_carrier_date', 'review_answer_timestamp', 'review_creation_date', 'shipping_limit_date'], inplace= True)
```

```python
Total = df.isnull().sum().sort_values(ascending = False)          
Percent = (df.isnull().sum()*100/df.isnull().count()).sort_values(ascending = False)   
missing_data = pd.concat([Total, Percent], axis = 1, keys = ['Total', 'Percentage of Missing Values'])    
missing_data
```

### Convert Into Time-Stamp

```python
column_time = ['order_purchase_timestamp','order_approved_at', 'order_delivered_customer_date', 'order_delivered_carrier_date', 
               'review_answer_timestamp', 'review_creation_date', 'shipping_limit_date']
for col in column_time:
    df[col]=pd.to_datetime(df[col])
```

```python
df['order_purchase_date']=pd.to_datetime(df['order_purchase_timestamp'])

df['order_purchase_date'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_purchase_date'] = df['order_purchase_date'].dt.strftime('%Y-%m-%d-%H')
df['order_purchase_date'] = pd.to_datetime(df['order_purchase_date'])

df['year'] = df['order_purchase_date'].dt.year
df['month'] = df['order_purchase_date'].dt.strftime('%B')
df['day'] = df['order_purchase_date'].dt.day
df['hour'] = df['order_purchase_date'].dt.hour
```

<a id="3.10"></a>
<font color="dimgrey" size=+2.0><b>3.10 Comparative Brazilian Ecommerce by Time</b></font>

```python
df['ord_purchase_dayofweek'] = df['order_purchase_timestamp'].apply(lambda x: x.strftime('%a'))
```

```python
fig, (ax1,ax2)=plt.subplots(1,2, figsize=(10,5))
plt.suptitle('Comparative Brazilian Ecommerce by Time', fontsize= 15)

df['hour']=df['hour'].astype(int)
hours_bins = [-0.1, 6, 12, 18, 23]
hours_labels = ['Dawn', 'Morning', 'Afternoon', 'Night']
df['order_purchase_time_day'] = pd.cut(df['hour'], hours_bins, labels = hours_labels)
sns.countplot(data= df, x='order_purchase_time_day', 
              ax= ax1, order= df['order_purchase_time_day'].value_counts().index, palette='inferno')
ax1.set_ylabel('Count')
ax1.set_xlabel('Time of Day')
ax1.set_title('Total Order by Time of Day')

sns.countplot( x='ord_purchase_dayofweek',data= df, ax= ax2,
              palette='YlGnBu', order= df['ord_purchase_dayofweek'].value_counts().index)
ax2.set_ylabel(ylabel= None)
ax2.set_xlabel('Order by Day of Week')
ax2.set_title('Total Order by Day of Week')
plt.tight_layout()
plt.show()
```

<h3 style='text-align: center;'>Insights</h3>

- Customer paling banyak membeli pada siang-sore hari
- Customer paling banyak membeli di hari Senin - Selasa, hal ini disebabkan karena pada hari Sabtu dan Minggu, Customer menghabiskan waktu untuk hiburan dan keluarga

<font color="dimgrey" size=+2.0><b>3.12 Population Map Distribution</b></font>
